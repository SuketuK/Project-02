```python
# clean Age data from lusitania data set

pattern_brackets_1 = '\(..\?\)'
pattern_brackets_2 = '\(.\?\)'
pattern_q = ' \(\?\)'
pattern_q2 = '\?'
pattern_months = '..-months'
pattern_infant = 'Infant'
pattern_spaces = ' '
pattern_underscore = '_'
pattern_exception1 = 'or..'

lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_brackets_1,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_brackets_2,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_q,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_q2,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_months,'1')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_infant,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_spaces,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_underscore,'')
lusitania_data['Age']=lusitania_data['Age'].str.replace(pattern_exception1,'')
lusitania_data['Age'].unique()
```

#### 1) Titanic summary by pclass, sex, embarked and status (deceased, survived)

#### Table below represents a pivot table by pclass, sex, embarked and status


```python
# titanic summary by pclass, sex, embarked and status
titanic_data_summary = titanic_data.pivot_table('age',
                                      columns=['status'],
                                      index=['pclass','sex','embarked'],
                                      aggfunc='count')
titanic_data_summary['All']=\
                                    titanic_data_summary['survived']\
                                    +titanic_data_summary['deceased']
titanic_data_summary['SurvivalRate']=\
                                    titanic_data_summary['survived']\
                                    /titanic_data_summary['All']
titanic_data_summary['DeathRate']=\
                                    titanic_data_summary['deceased']\
                                    /titanic_data_summary['All']
titanic_data_summary
```

```python
print 'Total Deceased: ', titanic_data_summary['deceased'].sum()
print 'Total Survived:', titanic_data_summary['survived'].sum()
```

```python
# Lusitania summary by sex, passenger/crew, department and fate
lusitania_summary_by_sex_pass_dept_fate = lusitania_data.pivot_table('Value',
                                      columns=['Fate'],
                                      index=['Sex','Passenger/Crew','Department/Class'],
                                      aggfunc='count')
lusitania_summary_by_sex_pass_dept_fate['All']=\
                                    lusitania_summary_by_sex_pass_dept_fate['Lost']\
                                    +lusitania_summary_by_sex_pass_dept_fate['Saved']\
                                    +lusitania_summary_by_sex_pass_dept_fate['Saved (died from trauma)']\
                                    .apply(lambda x: 0 if pd.isnull(x) else x)

lusitania_summary_by_sex_pass_dept_fate['SurvivalRate']=\
                                    lusitania_summary_by_sex_pass_dept_fate['Saved']\
                                    /lusitania_summary_by_sex_pass_dept_fate['All']

lusitania_summary_by_sex_pass_dept_fate['DeathRate']=\
                                    (lusitania_summary_by_sex_pass_dept_fate['Lost']\
                                     +lusitania_summary_by_sex_pass_dept_fate['Saved (died from trauma)']\
                                    .apply(lambda x: 0 if pd.isnull(x) else x))\
                                    /lusitania_summary_by_sex_pass_dept_fate['All']

lusitania_summary_by_sex_pass_dept_fate
```
